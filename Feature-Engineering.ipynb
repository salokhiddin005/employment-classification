```python
df.index = df.index + 1
print(df)
```

```python
df['Years_Of_Experience'] = df['Years_Of_Experience'].astype(int)
df['Monthly_Salary_(Inr)'] = df['Monthly_Salary_(Inr)'].astype(int)
df
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for col in ['Status', 'Education', 'Industry', 
            'Location', 'Ai_Risk', 'Age_Group']:
    df[col] = le.fit_transform(df[col])
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df['Years_Of_Experience'] = pd.to_numeric(df['Years_Of_Experience'], errors='coerce')
df['Monthly_Salary_(Inr)'] = pd.to_numeric(df['Monthly_Salary_(Inr)'], errors='coerce')

numeric_df = df.select_dtypes(include=['int64', 'float64'])

corr_matrix = numeric_df.corr()

# Plot
plt.figure()
plt.imshow(corr_matrix)
plt.title("Correlation Matrix (Numeric Columns Only)")
plt.xticks(range(len(corr_matrix.columns)), corr_matrix.columns, rotation=45)
plt.yticks(range(len(corr_matrix.columns)), corr_matrix.columns)
plt.colorbar()
plt.show()
```
