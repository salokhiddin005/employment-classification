```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
counts = df['Status'].value_counts()

plt.figure()
plt.bar(counts.index, counts.values)
plt.title("Employment Status Count")
plt.xlabel("Status")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.show()
```

```python
risk_counts = df['Ai_Risk'].value_counts()

plt.figure()
plt.pie(risk_counts.values, labels=risk_counts.index, autopct='%1.1f%%')
plt.title("AI Risk Distribution")
plt.show()
```

```python
df['Date_Recorded'] = pd.to_datetime(df['Date_Recorded'], errors='coerce')

monthly_avg = (
    df.groupby(pd.Grouper(key='Date_Recorded', freq='ME'))
      ['Monthly_Salary_(Inr)']
      .mean()
)

plt.figure()
plt.plot(monthly_avg.index, monthly_avg.values)
plt.title("Average Monthly Salary Over Time")
plt.xlabel("Date")
plt.ylabel("Average Salary (INR)")
plt.xticks(rotation=45)
plt.show()
```

```python
top_ind = (df.dropna(subset=['Industry', 'Monthly_Salary_(Inr)'])
             .groupby('Industry')['Monthly_Salary_(Inr)']
             .mean()
             .sort_values(ascending=False)
             .head(10))

plt.figure()
plt.bar(top_ind.index, top_ind.values)
plt.title("Top 10 Industries by Avg Salary")
plt.xlabel("Industry")
plt.ylabel("Avg Salary (INR)")
plt.xticks(rotation=45, ha='right')
plt.show()
```
